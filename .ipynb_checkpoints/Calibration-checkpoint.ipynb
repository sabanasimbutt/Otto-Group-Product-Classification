```python
import numpy as np 
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import seaborn as sns; sns.set() 
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, log_loss
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv("C://Users/Saba Naseem/Desktop/kaggle/Otto Group Product Classification/train.csv")
print(data.shape)
data.head()
```

```python
def downcast(df):
    
    columns= [f for f in df if df[f].dtypes=='int64']
    df[columns]= df[columns].astype(np.int32)
    return df
```

```python
data= downcast(data)
```

```python
data['target'].unique()
```

```python
data.loc[data['target']=='Class_1', 'target']=1
data.loc[data['target']=='Class_2', 'target']=2
data.loc[data['target']=='Class_3', 'target']=3
data.loc[data['target']=='Class_4', 'target']=4
data.loc[data['target']=='Class_5', 'target']=5
data.loc[data['target']=='Class_6', 'target']=6
data.loc[data['target']=='Class_7', 'target']=7
data.loc[data['target']=='Class_8', 'target']=8
data.loc[data['target']=='Class_9', 'target']=9
data['target']= data['target'].astype(np.int32)
```

```python
data['target'].unique()
```

```python
plt.figure(figsize=[12,6])
sns.countplot(data['target'])
```

```python
data.drop('id', axis=1, inplace=True)
y=np.asarray(data['target'])
data.drop('target', axis=1, inplace=True)
x= np.asarray(data)
x.shape, y.shape
```

```python
xtrain, xtest, ytrain, ytest= train_test_split(x, y, test_size=0.3, random_state=4)
```

```python
my_model = RandomForestClassifier(random_state=0).fit(xtrain, ytrain)
```

```python
import eli5
from eli5.sklearn import PermutationImportance

perm = PermutationImportance(my_model, random_state=1).fit(xtest, ytest)
```

```python
eli5.show_weights(perm, feature_names = data.columns.tolist())
```

```python
plt.figure(figsize=[20,8])
plt.plot(perm.feature_importances_)
plt.xticks(np.arange(data.shape[1]), data.columns.tolist(), rotation=90);
```

```python
names=['feat_60','feat_34','feat_11','feat_26','feat_25','feat_14','feat_86','feat_67','feat_15','feat_42',
       'feat_40','feat_48','feat_62',
      'feat_88','feat_36','feat_90','feat_24','feat_43','feat_60','feat_68']
import time
```

```python
ts= time.time()
RF= RandomForestClassifier(n_estimators=100,criterion='entropy',max_features=30, random_state=0)

RF.fit(xtrain, ytrain)

predict= RF.predict(xtrain)
print("In-Sample accuracy:",accuracy_score(ytrain, predict))
predict= RF.predict_proba(xtrain)
print("Out-Sample logloss: ",log_loss(ytrain, predict))

predict= RF.predict(xtest)
print("Out-Sample accuracy: ",accuracy_score(ytest, predict))
predict= RF.predict_proba(xtest)
print("Out-Sample Logloss: ",log_loss(ytest, predict))
print("Time:",(time.time()-ts)/60)
```

```python
from sklearn.calibration import CalibratedClassifierCV
```

```python
calibrated_clf = CalibratedClassifierCV(RF, method='isotonic', cv=5)
calibrated_clf.fit(xtrain, ytrain)

predict= calibrated_clf.predict(xtrain)
print("In-Sample accuracy:",accuracy_score(ytrain, predict))
predict= calibrated_clf.predict_proba(xtrain)
print("Out-Sample logloss: ",log_loss(ytrain, predict))

predict= calibrated_clf.predict(xtest)
print("Out-Sample accuracy: ",accuracy_score(ytest, predict))
predict= calibrated_clf.predict_proba(xtest)
print("Out-Sample Logloss: ",log_loss(ytest, predict))
```

```python
RF= RandomForestClassifier(n_estimators=100,criterion='entropy',max_features=30, random_state=0)
RF.fit(x, y)
calibrated_clf = CalibratedClassifierCV(RF, method='isotonic', cv=5)
calibrated_clf.fit(x, y)
```

```python
test=pd.read_csv("C://Users/Saba Naseem/Desktop/kaggle/Otto Group Product Classification/test.csv")
idd=np.asarray(test['id'])
test.drop('id', axis=1,inplace=True)
xx=np.asarray(test)

predict= calibrated_clf.predict_proba(xx)
```

```python
a=pd.Series(idd)
array = pd.DataFrame(predict,columns=["Class_1","Class_2","Class_3","Class_4",
                                      "Class_5","Class_6","Class_7","Class_8","Class_9"])

final= pd.DataFrame(a,columns=['id'])

final["Class_1"]=array["Class_1"]
final["Class_2"]=array["Class_2"]
final["Class_3"]=array["Class_3"]
final["Class_4"]=array["Class_4"]
final["Class_5"]=array["Class_5"]
final["Class_6"]=array["Class_6"]
final["Class_7"]=array["Class_7"]
final["Class_8"]=array["Class_8"]
final["Class_9"]=array["Class_9"]

final.to_csv("submit12.csv",index=None)
```
